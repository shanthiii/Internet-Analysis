```python
dictionary = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
with open('statistics.log') as f:
  lines = f.readline()
  x = lines[0:3]
  d = {}
  count = 0
  while(x in dictionary):
    count += 1
    dict={}
    dict['day'] = x
    n = lines.strip().split(":")
    m = len(n[0])
    # print(v[0][m-2:],v[1],v[2][0:2])
    dict['time'] = n[0][m-2:]+":" +n[1]+":"+n[2][0:2]
    lines =  f.readline()
    # print(line_1)
    n = lines.strip().split(":")
    # print("hello",n)
    # while(n[0:3] in dictionary):    
    #   dict["Server"]=''
    #   dict['Latency'] = ''
    #   dict['Jitter'] = ''
    #   dict['Download'] = ''
    #   dict['Upload'] = ''
    #   dict['Packet Loss'] = ''
    #   print(count,d[count])
    #   n = line_1.strip().split(":")
    if(n[0]=="Server"):
      l = n[1].strip().split("(")
      # print(l[0].strip())
      dict["Server"] = l[0]
      lines =  f.readline()
      # print(line_2)
      n = lines.strip().split(":")
    if(n[0]=="ISP"):
      lines = f.readline()
      n = lines.strip().split(":")
    if(n[0]=="Latency"):
      l = n[1].strip().split("(")
      b = l[0].strip().split("ms")
      dict["Latency"] = float(b[0].strip())
      m = l[1].strip().split("ms")
      dict["Jitter"] = float(m[0].strip())
      lines =  f.readline()
        # print(line_3)
      n = lines.strip().split(":")
    if(n[0]=="Latency"):
      lines =  f.readline()
        # print(line_3)
      # n = lines.strip().split(":")
    if(n[0]=="Download"):
      l = n[1].strip().split("Mbps")
      if(l[0]!="FAILED"):
        dict['Download'] = float(l[0].strip())
      lines =  f.readline()
      n = lines.strip().split(":")
    # print(line_4,n)
    if(n[0]=="Download"):
      lines =  f.readline()
    if(n[0]=="Upload"):
      l = n[1].strip().split("Mbps")
      if(l[0]!="FAILED"):
        dict["Upload"] = float(l[0].strip())
      # print(l)
      lines =  f.readline()
      n = lines.strip().split(":")
    if(n[0]=="Packet Loss"):
      x = n[1].strip().split("%")
      if x[0] == "Not available.":
        lines = f.readline()
      else:
        dict["Packet Loss"] = float(x[0])
        lines = f.readline()
      # y = n
    d[count] = dict
    x = lines[0:3]
    # for i in range(10):
    print(count,d[count],x)
```
